```python
import numpy as np
import pandas as pd
import matplotlib_inline as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  accuracy_score
```

```python
data = pd.read_csv('/content/data.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
# now removing the unnamed col
data.drop('Unnamed: 32', axis=1, inplace=True, errors='ignore')
data.drop('id', axis=1, inplace=True, errors='ignore')
data.head()
```

```python
data.groupby('diagnosis').mean()
```

dividing the data

```python
x = data.drop(columns='diagnosis',axis=1)
y = data['diagnosis']
```

```python
print(x)
print(y)
```

trainig the data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x.shape,x_train.shape,x_test.shape
```

```python
print(x_train)
print(y_train)
```

 # Now applying Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy score is ', training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score is ', testing_data_accuracy)
```
